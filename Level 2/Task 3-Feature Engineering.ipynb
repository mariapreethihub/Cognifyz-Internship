```python
for col in df.select_dtypes('object'):                                   # For removing special characters
    df[col]=df[col].str.replace(r'[^a-zA-Z0-9,()]+',"",regex=True)
```

```python
df.isnull().sum()                                         # Checking for null values
```

```python
df['Cuisines']=df['Cuisines'].fillna(df['Cuisines'].mode()[0])        # Imputing null values in field 'Cuisines' with mode
```

```python
df.isnull().sum()                                                     # Rechecking for null values
```

```python
df.duplicated().sum()                                # Check for duplicate rows
```

```python
df['Cuisines']=df['Cuisines'].str.split(',')         # Field cuisines has multiple value stored as single string, same is split using split function.
df=df.explode('Cuisines')                            # Separate row for each cuisine is created using explode function
df.head()
```

## 4. Feature Engineering

```python
df.drop(columns='Switch to order menu',inplace=True)                # Dropping field 'Switch to order menu' as it does not contribute anything towards analysis
```

```python
df.rename(columns={'Aggregate rating':'Target'},inplace=True)      # Aggregate rating the target varibale is renamed as Target for better readability.
df.head()
```

```python
df.columns
```

```python
df['Name_Length'] = df['Restaurant Name'].str.len()          # New column Name_Length is created that displays length of restaurant name
```

```python
df.head()
```

```python
df['Address_length']=df['Address'].str.len()           # New column Address_length is created that has length of address
df.head()
```
